```python
import folium
import pandas as pd


# Cargar un GeoJSON de provincias de España 
geo_json_data = "/home/bosser/Documentos/PROYECTOFINAL/Graficas/data/municipios.geojson"

# Crear un mapa centrado en España
m = folium.Map(location=[40.4168, -3.7038], zoom_start=6)

# Añadir el mapa coroplético
folium.Choropleth(
    geo_data=geo_json_data,  # Archivo GeoJSON
    name="choropleth",
    data= manfred_cuenta,  # DataFrame con los datos
    columns=["Presencial", "size"],  # Columnas con el nombre de la región y el valor
    key_on="feature.properties.NAMEUNIT",  # Ruta en el GeoJSON para hacer el match con los nombres
    fill_color="Blues",  # Esquema de colores
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Tamaño",  # Leyenda
).add_to(m)

# Añadir un control de capas
folium.LayerControl().add_to(m)

# Mostrar el mapa en un notebook (si estás usando Jupyter)
m
```

```python
import json

# Función para capitalizar y eliminar signos de puntuación
def formatear_nombre(nombre):
    # Capitalizar cada palabra
    return nombre.title()

# Cargar el archivo GeoJSON
with open("/home/bosser/Documentos/PROYECTOFINAL/Graficas/data/municipios.geojson", "r", encoding="utf-8") as file:
    geojson_data = json.load(file)

# Modificar in-place los nombres de las provincias
for feature in geojson_data["features"]:
    feature["properties"]["nombre"] = formatear_nombre(feature["properties"]["nombre"])

# Guardar el GeoJSON modificado (opcional)
with open("/home/bosser/Documentos/PROYECTOFINAL/Graficas/data/municipios_formateados.geojson", "w", encoding="utf-8") as file:
    json.dump(geojson_data, file, ensure_ascii=False, indent=4)

# Imprimir los nombres formateados (para verificación)
print("Nombres de las provincias en el GeoJSON (formateados):")
for feature in geojson_data["features"]:
    print(feature["properties"]["nombre"])
    
```

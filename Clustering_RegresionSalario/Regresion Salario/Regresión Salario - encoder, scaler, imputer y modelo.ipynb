```python
#Queries para obtener el df general + top 10 tecnologias y top 5 habilidades.

config = toml.load("../.streamlit/secrets.toml")
db_config = config["database"]

db = mysql.connector.connect(host=db_config["host"],
                             user=db_config["user"],
                             password=db_config["password"],
                             database=db_config["database"]) 
query = """
WITH TopTecnologias AS (
    SELECT tr.tec_id, t.tecnologia
    FROM tecnologias_relacion tr
    JOIN tecnologias t ON tr.tec_id = t.tec_id
    GROUP BY tr.tec_id
    ORDER BY COUNT(*) DESC
    LIMIT 10),
TopHabilidades AS (
    SELECT hr.hab_id, h.habilidad
    FROM habilidades_relacion hr
    JOIN habilidades h ON hr.hab_id = h.hab_id
    GROUP BY hr.hab_id
    ORDER BY COUNT(*) DESC
    LIMIT 5),
TecnologiasOferta AS (
    SELECT 
        tr.id_oferta,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 0) THEN 1 ELSE 0 END) AS `Tecnologia_1`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 1) THEN 1 ELSE 0 END) AS `Tecnologia_2`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 2) THEN 1 ELSE 0 END) AS `Tecnologia_3`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 3) THEN 1 ELSE 0 END) AS `Tecnologia_4`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 4) THEN 1 ELSE 0 END) AS `Tecnologia_5`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 5) THEN 1 ELSE 0 END) AS `Tecnologia_6`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 6) THEN 1 ELSE 0 END) AS `Tecnologia_7`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 7) THEN 1 ELSE 0 END) AS `Tecnologia_8`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 8) THEN 1 ELSE 0 END) AS `Tecnologia_9`,
        MAX(CASE WHEN tr.tec_id = (SELECT tec_id FROM TopTecnologias LIMIT 1 OFFSET 9) THEN 1 ELSE 0 END) AS `Tecnologia_10`
    FROM tecnologias_relacion tr
    GROUP BY tr.id_oferta),
HabilidadesOferta AS (
    SELECT 
        hr.id_oferta,
        MAX(CASE WHEN hr.hab_id = (SELECT hab_id FROM TopHabilidades LIMIT 1 OFFSET 0) THEN 1 ELSE 0 END) AS `Habilidad_1`,
        MAX(CASE WHEN hr.hab_id = (SELECT hab_id FROM TopHabilidades LIMIT 1 OFFSET 1) THEN 1 ELSE 0 END) AS `Habilidad_2`,
        MAX(CASE WHEN hr.hab_id = (SELECT hab_id FROM TopHabilidades LIMIT 1 OFFSET 2) THEN 1 ELSE 0 END) AS `Habilidad_3`,
        MAX(CASE WHEN hr.hab_id = (SELECT hab_id FROM TopHabilidades LIMIT 1 OFFSET 3) THEN 1 ELSE 0 END) AS `Habilidad_4`,
        MAX(CASE WHEN hr.hab_id = (SELECT hab_id FROM TopHabilidades LIMIT 1 OFFSET 4) THEN 1 ELSE 0 END) AS `Habilidad_5`
    FROM habilidades_relacion hr
    GROUP BY hr.id_oferta),
PrimerIdioma AS (
    SELECT 
        id_oferta,
        MIN(idioma) AS PrimerIdioma
    FROM idiomas
    GROUP BY id_oferta),
PrimeraCiudad AS (
    SELECT 
        id_oferta,
        MIN(ciudad) AS PrimeraCiudad
    FROM ciudades
    GROUP BY id_oferta)
SELECT 
    (o.salario_desde + o.salario_hasta) / 2 AS salario,
    o.variabe,
    o.vacaciones,
    o.teletrabajo,
    o.jornada_tipo,
    o.funciones,
    o.contrato_tipo,
    o.nivel_profesional,
    o.formacion_minima,
    o.personas_a_cargo,
    o.experiencia,
    pi.PrimerIdioma,
    pc.PrimeraCiudad,
    t.Tecnologia_1,
    t.Tecnologia_2,
    t.Tecnologia_3,
    t.Tecnologia_4,
    t.Tecnologia_5,
    t.Tecnologia_6,
    t.Tecnologia_7,
    t.Tecnologia_8,
    t.Tecnologia_9,
    t.Tecnologia_10,
    h.Habilidad_1,
    h.Habilidad_2,
    h.Habilidad_3,
    h.Habilidad_4,
    h.Habilidad_5
FROM 
    ofertas o
LEFT JOIN 
    PrimerIdioma pi ON o.id_oferta = pi.id_oferta
LEFT JOIN 
    PrimeraCiudad pc ON o.id_oferta = pc.id_oferta
LEFT JOIN 
    TecnologiasOferta t ON o.id_oferta = t.id_oferta
LEFT JOIN 
    HabilidadesOferta h ON o.id_oferta = h.id_oferta;
"""
query_top10tec = """
SELECT t.tecnologia
    FROM tecnologias_relacion tr
    JOIN tecnologias t ON tr.tec_id = t.tec_id
    GROUP BY tr.tec_id
    ORDER BY COUNT(*) DESC
    LIMIT 10;
"""
query_top5hab = """
    SELECT h.habilidad
    FROM habilidades_relacion hr
    JOIN habilidades h ON hr.hab_id = h.hab_id
    GROUP BY hr.hab_id
    ORDER BY COUNT(*) DESC
    LIMIT 5;
"""
df_general = pd.read_sql(query, db)

top10tec_df = pd.read_sql(query_top10tec, db)
top10tec = top10tec_df["tecnologia"].tolist()
top5hab_df = pd.read_sql(query_top5hab, db)
top5hab = top5hab_df["habilidad"].tolist()

db.close()
```

```python
#Usamos el encoder para codificar el df

df_codificado = df_general.copy()
def transform_column(df, column_name, encoder):
    if column_name in encoder:
        mapping = encoder[column_name]
        if mapping is None:
            return df[column_name]
        else:
            return df[column_name].replace(mapping).fillna(df[column_name])
    else:
        return df[column_name]
for column in df_codificado.columns:
    df_codificado[column] = transform_column(df_codificado, column, encoder)
```

```python
#Escalamos el df con MinMaxScaler

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_codificado)  # Esto devuelve un numpy array
# Convertir el resultado a un df
df_scaled = pd.DataFrame(df_scaled, columns=df_codificado.columns)
```

```python
#Imputamos valores faltantes con KNNImputer

imputer = KNNImputer(n_neighbors=5)
columnas = df_scaled.columns
df_imputado = imputer.fit_transform(df_scaled)
df_imputado = pd.DataFrame(df_imputado, columns = df_scaled.columns)
df_imputado.to_csv("df_imputado_regresion_salario", index = False)
```

```python
#Con la grilla de hiperparámetros, escogemos el mejor modelo posible.

X = df_imputado.drop("salario", axis=1)
y = df_imputado["salario"] #Separamos la columna objetivo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

model = RandomForestRegressor(random_state=42)

# Grilla de hiperparámetros
param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"]}

# Configurar Grid Search
grid_search = GridSearchCV(estimator=model,
                           param_grid=param_grid,
                           scoring="neg_mean_squared_error",
                           cv=5,
                           n_jobs=-1)

# Ajustar Grid Search a los datos de entrenamiento
grid_search.fit(X_train, y_train)

# Obtener los mejores hiperparámetros
best_params = grid_search.best_params_
print("Mejores hiperparámetros:", best_params)

# Evaluar el mejor modelo en el conjunto de prueba
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"MSE con el mejor modelo: {mse}")
```

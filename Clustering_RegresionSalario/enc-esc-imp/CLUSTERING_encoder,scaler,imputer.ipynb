```python
df_numeros.rename(columns={"id_oferta": "id"}, inplace=True)
df_general.rename(columns={"id_oferta": "id"}, inplace=True)
df_general = pd.merge(df_general, df_numeros, on="id", how="left")
df_general["salario_media"] = (df_general["salario_desde"] + df_general["salario_hasta"]) / 2
df_general["variable_booleano"] = ~df_general["variabe"].isna() & (df_general["variabe"] != 0)
columna_id = df_general["id"]
df_general = df_general.drop(columns=["ciudad_x", "id", "empresa", 
                                      "titulo", "url", "timestamp", 
                                      "fecha_publicacion", "oferta_activada", 
                                      "cvs_inscritos", "imprescindible_residir", 
                                      "vacantes"])
df_general.rename(columns={"variabe": "variable"}, inplace=True)
df_general.rename(columns={"ciudad_y": "ciudad"}, inplace=True)
df_general["ciudad"] = df_general["ciudad"].str.title()
df_general = df_general.drop(columns=["salario_desde", "salario_hasta"])
col = df_general.pop("salario_media")
df_general.insert(4, "salario_media", col)
```

```python
#Creamos encoder para transformar todas las columnas categóricas a numéricas.

encoder = {
    "salario_media": None,
    "variable": lambda df: df["variable"].fillna(0),
    "variable_booleano": lambda df: df["variable"].apply(lambda x: 1 if x != 0 else 0),
    "vacaciones": lambda df: df["vacaciones"].mode()[0] if not df["vacaciones"].mode().empty else 0,
    "teletrabajo": {"Presencial": 0, "Hibrido": 1, "Remoto": 2},
    "jornada_tipo": {"Por Horas": 1,
                     "Media Jornada": 2,
                     "Turno Rotatorio": 3,
                     "Intensiva Tarde": 4,
                     "Jornada Completa": 4.1,
                     "Intensiva Mañana": 4.3,
                     "Jornada Completa Flexible": 4.4},

    "turno_continuo": None,
    "funciones": lambda df: df["funciones"].map({cat: i for i, cat in enumerate(df["funciones"].dropna().unique())}),
    "contrato_tipo": {"A determinar": 1,
                      "Prácticas": 2,
                      "Obra o servicio": 3,
                      "Temporal": 4,
                      "Freelance/Autónomo": 5,
                      "Indefinido": 6},
    "nivel_profesional": {"Prácticas / Beca": 1,
                          "Empleado": 5,
                          "Mando Intermedio": 20,
                          "Especialista": 30,
                          "Director / Gerente": 40,
                          "Administrador / Consejero": 50},
    "formacion_minima": {"Sin estudios": 15,
                         "E.S.O. (Educación Secundaria Obligatoria)": 16,
                         "FP1": 18,
                         "Grado Medio": 18,
                         "Bachillerato/COU": 18.5,
                         "Certificado de Profesionalidad": 19,
                         "Otros títulos, certificaciones y carnets": 19.4,
                         "Otra Formación Tecnológica": 19.8,
                         "FP2/Grado Superior": 20,
                         "Diplomado": 21,
                         "Ingeniero Técnico": 21.5,
                         "Grado EEES (Bolonia)": 22,
                         "Licenciado": 23,
                         "Ingeniero Superior": 23.5,
                         "Postgrado EEES (Máster)": 24,
                         "Doctorado": 28},
    "personas_a_cargo": lambda df: df["personas_a_cargo"].fillna(0),
    "experiencia": {"Sin Experiencia": 1,
                    "Menos de un año": 2,
                    "1 año": 2,
                    "2 años": 5,
                    "3 años": 5,
                    "3-5 años": 5,
                    "default": 10},
    "num_tecnologias": None,
    "num_habilidades": None,
    "ciudad": lambda df: df["ciudad"].map({cat: i for i, cat in enumerate(df["ciudad"].dropna().unique())})}
```

```python
#Codificamos el df

for columna, transformacion in encoder.items():
    if columna not in df_general.columns:
        print(f"Columna {columna} no encontrada en el DataFrame, omitiendo transformación.")
        continue
    if transformacion is None:
        continue
    elif callable(transformacion):
        df_general[columna] = transformacion(df_general)
    elif isinstance(transformacion, dict):
        df_general[columna] = df_general[columna].map(transformacion)
    else:
        df_general[columna] = transformacion  # En caso de que sea un valor constante
```

```python
#Escalamos los datos con MinMaxScaler

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_general)  # Esto devuelve un numpy array
df_scaled = pd.DataFrame(df_scaled, columns=df_general.columns) # Convertimos a df
```

```python
#Imputamos datos faltantes. Y guardamos en csv el df imputado con el id de nuevo.

imputer = KNNImputer(n_neighbors=3)
df_imputado = imputer.fit_transform(df_scaled)
df_imputado = pd.DataFrame(df_imputado, columns=df_scaled.columns)
df_imputado["id"] = columna_id
df_imputado.to_csv("df_imputado_clustering.csv", index=False)
```

```python
#Guardamos el encoder, el scaler y el imputer en pickles

with open("encoder_clustering.pkl", "wb") as f:
    dill.dump(encoder, f)

with open("scaler_clustering.pkl", "wb") as f:
    pickle.dump(scaler, f)

with open("imputer_clustering.pkl", "wb") as f:
    pickle.dump(imputer, f)
```

```python
#Renombramos columnas en los df
df_tecnoempleo = df_tecnoempleo.rename(columns={"Título": "titulo", "Nombre de empresa": "empresa", "Ciudad": "ciudad", "Salario_desde": "salario_desde", "Salario_hasta": "salario_hasta", "Jornada": "jornada_tipo", "Teletrabajo": "teletrabajo", "Time Stamp": "timestamp", "Fecha publicacion" : "fecha_publicacion", "Imprescindible Residir" : "imprescindible_residir", "Funciones" : "funciones", "Nivel Profesional" : "nivel_profesional", "CVs inscritos en el proceso" : "cvs_inscritos", "Otras Provincias" : "otras_provincias", "Número de puestos" : "vacantes", "Personas a cargo" : "personas_a_cargo", "+ Países" : "otros_paises", "URL" : "url", "Experiencia" : "experiencia", "Tipo contrato" : "contrato_tipo", "Formación Mínima" : "formacion_minima"})
df_manfred = df_manfred.rename(columns={"Presencial": "ciudad", "Salario_desde": "salario_desde", "Salario_hasta": "salario_hasta", "Jornada": "jornada_tipo", "Teletrabajo": "teletrabajo", "fecha_extraccion" : "timestamp"})
```

```python
#La columna incentivos en principio no debería estar. Más adelante, eliminar esta línea

df_tecnoempleo.drop(columns = ["Incentivos", "Info Otros Países", "otros_paises"], inplace = True)
```

```python
df_fusion = pd.concat([df_tecnoempleo, df_manfred], axis=0, join='outer', ignore_index=True)
```

```python
orden_columnas_fusion = ["id","empresa","titulo","ciudad", "salario_desde", "salario_hasta","variable", "vacaciones", "teletrabajo", "jornada_tipo", "turno_continuo", "imprescindible_residir", "funciones", "contrato_tipo", "nivel_profesional", "formacion_minima", "personas_a_cargo", "experiencia", "vacantes", "cvs_inscritos", "oferta_activada", "fecha_publicacion", "timestamp", "url"]
```

```python
df_fusion = df_fusion[orden_columnas_fusion]
```

```python
df_fusion
```

```python
df_fusion["teletrabajo"] =  df_fusion["teletrabajo"].apply(lambda x : "Remoto" if x == 100 or x == "Remoto" else "Presencial" if x == 0 or x == "Presencial" else "Hibrido")

df_fusion["jornada_tipo"] = df_fusion["jornada_tipo"].astype(str).str.title()
```

```python
#Pasamos las columnas fecha_publicacion y timestamp a datetime, 
#errors = coerce convierte los nan a NaT (not a time)
df_fusion["fecha_publicacion"] = pd.to_datetime(df_fusion["fecha_publicacion"], errors = "coerce")
df_fusion["timestamp"] = pd.to_datetime(df_fusion["timestamp"], errors = "coerce")
```

```python
df_fusion
```

```python
df_fusion["oferta_activada"] = df_fusion["oferta_activada"].fillna(1).astype(int)
```

```python
df_fusion.to_csv("../../../CSVs/CSV_finales_merged/ofertas_final.csv", index=False, encoding="utf-8-sig")
```

```python
def habilidades_relacion (csv = "../../../CSVs/CSV_finales_merged/habilidades_matricial_final.csv", habilidades="../../../CSVs/CSV_finales_merged/habilidades_final.csv", relacion="../../../CSVs/CSV_finales_merged/habilidades_relacion_final.csv"):
    # Ejemplo de DataFrame
    df = pd.read_csv(csv)
    
    # Lo convertimos hacia abajo para que sea mas sencillo
    df_largo = df.melt(id_vars=["id"], var_name="habilidades", value_name="presente")
    #Si es igual a 1 lo mantenemos
    df_largo = df_largo[df_largo["presente"] == 1].drop(columns=["presente"])
    
    #Creamos un df de tecnoligas poniendole un id 
    df_habilidades = pd.DataFrame({"habilidades": df_largo["habilidades"].unique()})
    df_habilidades["hab_id"] = range(1, len(df_habilidades) + 1)
    
    # Relacionamos los dos
    df_relacion = df_largo.merge(df_habilidades, on="habilidades", how="left").drop(columns=["habilidades"])


    df_habilidades.to_csv(habilidades, index = False)
    df_relacion.to_csv(relacion, index = False)

    return df_habilidades, df_relacion 
```

```python
def tecnologias_relacion (csv = "../../../CSVs/CSV_finales_merged/tecnologias_matricial_final.csv", tecnologias="../../../CSVs/CSV_finales_merged/tecnologias_final.csv", relacion="../../../CSVs/CSV_finales_merged/relacion_tecnologias_final.csv"):
    # Ejemplo de DataFrame
    df = pd.read_csv(csv)
    
    # Lo convertimos hacia abajo para que sea mas sencillo
    df_largo = df.melt(id_vars=["id"], var_name="tecnologia", value_name="presente")
    #Si es igual a 1 lo mantenemos
    df_largo = df_largo[df_largo["presente"] == 1].drop(columns=["presente"])
    
    #Creamos un df de tecnoligas poniendole un id 
    df_tecnologias = pd.DataFrame({"tecnologia": df_largo["tecnologia"].unique()})
    df_tecnologias["tec_id"] = range(1, len(df_tecnologias) + 1)
    
    # Relacionamos los dos
    df_relacion = df_largo.merge(df_tecnologias, on="tecnologia", how="left").drop(columns=["tecnologia"])


    df_tecnologias.to_csv(tecnologias, index = False)
    df_relacion.to_csv(relacion, index = False)
```
